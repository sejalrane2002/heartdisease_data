```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'C:\\Users\\Sejal\\Desktop\\data science projects\\heart disease\\Heart Disease data\\Heart Disease data.csv'
heart_disease_data = pd.read_csv(file_path)
```

```python
missing_values = heart_disease_data.isnull().sum()
print(missing_values)

# Get basic statistics of the dataset
basic_stats = heart_disease_data.describe()
print(basic_stats)
```

```python
# Set the style for the plots
sns.set(style="whitegrid")
```

```python
# Plot the distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(heart_disease_data['age'], bins=30, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of sex
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', data=heart_disease_data)
plt.title('Gender Distribution of Patients')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of target variable
plt.figure(figsize=(10, 6))
sns.countplot(x='target', data=heart_disease_data)
plt.title('Heart Disease Diagnosis')
plt.xlabel('Target (0 = No Disease, 1 = Disease)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the relationship between gender and heart disease
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', hue='target', data=heart_disease_data)
plt.title('Heart Disease by Gender')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Frequency')
plt.legend(title='Heart Disease', loc='upper right', labels=['No Disease', 'Disease'])
plt.show()
```

```python
# Plot the relationship between age and heart disease
plt.figure(figsize=(12, 8))
sns.histplot(data=heart_disease_data, x='age', hue='target', multiple='stack', kde=True)
plt.title('Heart Disease by Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
